```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

```

```python
sns.get_dataset_names()
```

```python
gyser=sns.load_dataset('geyser')
gyser.head()
```

```python
gyser.describe()
```

## seaborn data visulation

## 1. amount of gyser on type

```python
sns.countplot(x='kind',data=gyser)
```

 ### we notice long gysers is more puplic and common to use

### 2.  waiting depends on kind of gyser :

```python
# sns.set_style("whitegrid")

# sns.barplot(x='duration', y='waiting', data=gyser)
g = sns.FacetGrid(gyser, col="kind")
g.map(sns.scatterplot, "duration", "waiting", alpha=.7)
g.add_legend()
```

### We notice from the graph that the long gysers waiting time is more than short after same time of uasge .

### 3. waiting time vs duration of use with ignore of kind of gyser : 

```python
sns.relplot(x='duration', y='waiting', data=gyser)
```

### we notice from the graph that when duration of usage increase  the waiting increases

Conclusion:

1. we notice long gysers is more puplic and common to use.
2. We notice from the graph that the long gysers waiting time is more than short after same time of uasge .
3. we notice from the graph that when duration of usage increase  the waiting increases.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
 
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()

```

## seaborn data visulation 

### 1. Gender Stat:
 

```python
sns.countplot(x='sex',data=tips)
```

### We can notice the count of males in this place more than females . 

## 2.tip VS  AVG Bill amount 

```python
data_grouped_avg= tips.groupby('total_bill', as_index=False).mean()
data_grouped_avg
```

```python
sns.set_style("whitegrid")

sns.relplot(x='total_bill', y='tip', data=data_grouped_avg , kind='scatter')
```

### Generally We notice from the figure that when the amount total bill increases the tip increases .

### 3. Smoker male vs Female and tip Value 

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip", alpha=.7)
g.add_legend()
```

### we notice from the  graph the males are more smoker than females . and the tips for 
### non-smoker is better and the  tips doen't always on gender .
